### Import the necessary libraries and the data set

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
population_data = pd.read_csv('WorldBank_Population.csv')
```

### Display the data set

```python
population_data.shape
```

```python
population_data.head(5)
```

```python
# Drop Unnamed Column
population_data.drop("Unnamed: 68", axis=1)
```

### Extract the relevent columns

```python
data_2023 = population_data.loc[:, ['Country Name', '2023']]
```

### Remove Rows with Missing Values

```python
data_2023.dropna()
```

```python
data_2023.head()
```

### Visualize the data using a Histogram

```python
# Set up the visual style
sns.set(style="whitegrid")

# Histogram for population distribution in 2022
plt.figure(figsize = (12, 6))
sns.histplot(data = data_2023, x='2023', bins=30, kde=True, color='crimson')
plt.title('Population Distribution in 2023')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

### Visualize the data using a Bar Chart

```python
# Sort data by population in descending order and select top 220
top_20_countries = data_2023.sort_values(by='2023', ascending=False).head(20)

# Set up the visual style
sns.set(style="whitegrid")

# Bar chart for top 20 most populous countries in 2023
plt.figure(figsize=(20, 6))
sns.barplot(data = top_20_countries, x='Country Name', y='2023', palette='gist_heat')
plt.title('Top 20 Most Populous Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
